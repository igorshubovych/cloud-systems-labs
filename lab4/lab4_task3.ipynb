### Імпорт бібліотек і налаштування

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.interpolate import lagrange

sns.set_theme()
```

### Вхідні дані

```python
# Вхідні дані
X = np.array([0.1, 0.3, 0.4, 0.6, 0.7])
Y = np.array([3.2, 3.0, 1.0, 1.8, 1.9])


# Вхідні дані (варіант 8 із завдання 2)
# X = np.array([6, 7, 8, 9, 10, 12])
# Y = np.array([2, 3, 3, 4, 6, 5])
```

### Інтерполяція методом Лагранжа

```python
# Створення інтерполяційного поліному
polynomic_function = lagrange(X, Y)

print("Поліноміальна функція")
print("Y = ")
print(polynomic_function)
```

### Графік

```python
# Побудова графіка
x_range = np.linspace(min(X), max(X), 500)
y_pred = polynomic_function(x_range)

plt.figure()
plt.plot(x_range, y_pred, label='Поліноміальна функція', color='blue',
    linestyle='--')
plt.scatter(X, Y, color='red', label='Вхідні дані')
plt.title('Інтерполяція методом поліному Лагранжа')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.legend()
plt.show()
```

### Обчислення проміжних значень

```python
# Обчислення значень у точках 0.2 і 0.5
print("Значення функції в точці 0.2:", polynomic_function(0.2))
print("Значення функції в точці 0.5:", polynomic_function(0.5))
```
