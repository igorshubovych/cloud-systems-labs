## Завдання 2. 
Експериментально отримані N-значень величини Y при значеннях величини Х. Відшукати параметри функції за методом найменших квадратів. Побудувати графіки, де в декартовій системі координат нанести експериментальні точки і графік апроксимуючої функції.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

sns.set_theme()
```

```python
# Варіант 8
X = np.array([6, 7, 8, 9, 10, 12]).reshape(-1, 1)
Y = np.array([2, 3, 3, 4, 6, 5])
```

```python
# Побудова графіка
plt.figure()
plt.scatter(X, Y, color='red', label='Дані')
plt.title('Вхідні дані (Варіант 8)')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.legend()
plt.show()
```

```python
# Створення та навчання моделі лінійної регресії
model = LinearRegression()
model.fit(X, Y)
x_range = np.linspace(min(X), max(X), 100).reshape(-1, 1)
y_pred = model.predict(x_range)

# Отримання коефіцієнтів
coefficient_0 = model.intercept_
coefficient_1 = model.coef_[0]


# Побудова графіка
equation_string = 'y = {:.3f}*x + {:.3f}'.format(coefficient_1, coefficient_0)

plt.figure(figsize=(9, 8))
plt.scatter(X, Y, color='red', label='Дані')
plt.plot(x_range, y_pred, color='blue',
         linestyle='--',
         label='Лінійно-регресійна пряма: ' + equation_string)
plt.title('Лінійна регресія')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.figtext(.02, .02, "Знаходження коефіцієнтів лінійної функції методом лінійної регресії")
plt.show()

print(f"Коефіцієнт β0: {coefficient_0}")
print(f"Коефіцієнт β1: {coefficient_1}")
```
