Експериментально отримані N-значень величини Y при значеннях величини Х. Відшукати параметри функції за методом найменших квадратів. Побудувати графіки, де в декартовій системі координат нанести експериментальні точки і графік апроксимуючої функції.

### Імпорт бібліотек і налаштування

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

sns.set_theme()
```

### Вхідні дані

```python
# Варіант 8
X = np.array([6, 7, 8, 9, 10, 12]).reshape(-1, 1)
Y = np.array([2, 3, 3, 4, 6, 5])

# Вхідні дані на графіку
plt.figure()
plt.scatter(X, Y, color='red', label='Дані')
plt.title('Вхідні дані (Варіант 8)')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.legend()
plt.show()
```

### Лінійна регресія

```python
# Створення та навчання моделі лінійної регресії
model = LinearRegression()
model.fit(X, Y)
x_range = np.linspace(min(X), max(X), 100).reshape(-1, 1)
y_pred = model.predict(x_range)

# Отримання коефіцієнтів
coefficient_0 = model.intercept_
coefficient_1 = model.coef_[0]

# Формування рядка з рівнянням лінійної регресії
function_string = 'y = {:.3f}*x + {:.3f}'.format(coefficient_1, coefficient_0)
print("Апроксимуюча функція, створена методом лінійної регресії")
print(function_string)
print("-------------------------")

print(f"Коефіцієнт β0: {coefficient_0}")
print(f"Коефіцієнт β1: {coefficient_1}")
```

### Графік

```python

# Побудова графіка

plt.figure(figsize=(9, 8))
plt.scatter(X, Y, color='red', label='Дані')
plt.plot(x_range, y_pred, color='blue',
         linestyle='--',
         label='Пряма апроксимуючої функції: ' + function_string)
plt.title('Лінійна регресія')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.legend()
plt.figtext(.02, .02, "Знаходження коефіцієнтів апроксимуючої функції методом лінійної регресії")
plt.show()
```
